```python
#------------------------------------------------------------------------Tasks
 #Produce a chart showing TC and TR with Q on the horizontal axis. 
 #Have Q go from 0 to 10,000 units. 
 #Produce a second chart showing MC and MR with Q again on the horizontal axis.
 #What is the optimal level of output for your company to produce/sell? 
 #What are the total revenue, total cost, and profit at optimal number of units?
 
 #-------------------------------------------------------------------Equations
 #Total Cost: TC = 31,000 + 7Q + 0.002Q^2
 #Marginal Cost: MC = 7 + 0.004Q  
 #P and MR= $37
 #TR = $37Q
#Profit = TR-TC 
```

---

```python
#Create Vectors for TC,MC, P, MR, and Profit

#Range of Q values 
Qrange = range(0,10001)

#Price
P = 37

#Vectors  
TC =[]
for q in Qrange:
    c = 31000.000 + 7.000*q + 0.002*q**2 #Python's exponential operand is ** instead of ^ 
    TC.append(c)
    
MC = []
for q in Qrange:
    m = 7 + (0.004*q)
    MC.append(m) 
    
TR = []
for q in Qrange:
    r = P*q
    TR.append(r)
    
MR = []
counter = 0
while counter <10001:
    counter = counter + 1
    MR.append(37)

Profit = []
for q in Qrange:
    f = (P*q) - (31000 + 7*q + 0.002*(q**2)) 
    Profit.append(round(f,2))
            
```

```python
#Convert Vectors into a Data Frame 
import pandas as pd 
RC = {"Q":Qrange,"TC":TC,"MC":MC,"TR": TR,"Profit":Profit,"MR":MR}
RCtable = pd.DataFrame(RC)
```

```python
#Check how table looks
RCtable.tail()
```

```python
#Find optimum level of output
optimumfilter = RCtable.MC == RCtable.MR
RCtable[optimumfilter]
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#Plot TC and TR Curve
plt.plot(RCtable.Q, RCtable.TC, color = "red")
plt.plot(RCtable.Q, RCtable.TR, color = "green")
plt.title("Total Revenue v. Total Cost")
plt.ylabel("Dollars")
plt.figlegend(labels = ("TC","TR"), loc="upper left", bbox_to_anchor = (0.15,0.85)) 
plt.show()
```

```python
#Plot Marginal Revenue and Marginal Cost Curves 
plt.plot(RCtable.Q, RCtable.MC,color = "red")
plt.plot(RCtable.Q, RCtable.MR, color = "green")
plt.title("Marginal Revenue v. Marginal Cost")
plt.ylabel("Dollars")
plt.figlegend(labels = ("MC","MR"), loc="upper left", bbox_to_anchor = (0.09,0.85)) 
plt.show()
```

```python
#Profit Curve
plt.plot(RCtable.Q, RCtable.Profit,color = "green")
plt.title("Profit")
plt.ylabel("Dollars")
plt.show()
```
