Question

On April 1st, the Mk I solar panel generates 1kJ/day of power

On May 1st, the MkII solar panel generates 4 kJ/day of power

1. On what day does Mk II generate as much energy as MkI?
2. How much total energy have they generated by that day? 
3. What would the solutions to questions 1 and 2 be if Mark II generated 1kJ a day?

---

Solution

Mk I power: pI = t + 30 (30 is added to t because 30 days, thus 30kJ, will have passed by the time Mk II activates)

MkII power: pII = 4t  

    30 + t = 4t 
    
    30 = 3t
    
    t = 10

1. Thus, on the 10th day of Mk II's activation (40 days after MkI activates), both models will have generated the same amount of energy.
2. MkI generated 40 kJ and MkII generated 40kJ on May 10th.
3. If pII = t: 

    t + 30 = t
    
    thus, there is no solution because the lines will never meet

Plot solution

```python
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib.dates as mdates
from matplotlib.dates import DayLocator, HourLocator, DateFormatter, drange
import datetime as dt
import pandas as pd
import datetime as dt
```

```python
#define the time variable t representing days
t = np.linspace(-30,30,1000)
```

```python
#MkI power equation
# +30 because on May 1st, the starting point for MkII, MkI will have already generated 30 kJ
p_I= t +30
```

```python
#MkII power equation
p_II = 4*t
```

```python
##Plot equations to display the time and distance when their paths cross

#Plot Header and Axes Titles
fig,ax = plt.subplots()
plt.title("MkI and MkII Power Generation")
plt.xlabel('Days')
plt.ylabel('Power (kJ)')

#Plot Data
ax.set_xlim([-30,31])
ax.set_ylim([0,50])
ax.plot(t,p_I,c= 'brown')
ax.plot(t,p_II,c = 'green')

#Plot hash lines  
plt.axvline(x=10, color = 'purple',linestyle = '--')
_ = plt.axhline(y = 40, color = 'purple', linestyle = '--')







```
