# Run Multiple Linear Regressions

```python
#Model A: Robbery = f(%Metro, %Poverty, Rain)
#Model B: Robbery = f(Population, %SameHouse, MedianAge)
#Model C: Robbery = f(Population, %SameHouse, MedAgeMale)
#Model D: Robbery = f(Population, %SameHouse, MedAgeFemale)
#Model E: Robbery = f(CarTheft, %Unemploy, %Bachelors+, Temp)
#Model F: Robbery = f(CarTheft, %Unemploy, %Bachelors+, Sun)
#Model G: Robbery = f(CarTheft, %Unemploy, %Bachelors+, Rain)
```

```python
#Define Dependent and Idependent Variables
y = crimestats["Robbery"]
x1A = crimestats[["%Metro", "%Poverty", "Rain"]]
x1B = crimestats[["Population", "%SameHouse", "MedianAge"]]
x1C = crimestats[["Population", "%SameHouse", "MedAgeMale"]]
x1D = crimestats[["Population", "%SameHouse", "MedAgeFemale"]]
x1E = crimestats[["CarTheft", "%Unemploy", "%Bachelors+", "Temp"]]
x1F = crimestats[["CarTheft", "%Unemploy", "%Bachelors+", "Sun"]]
x1G = crimestats[["CarTheft", "%Unemploy", "%Bachelors+", "Rain"]] 
```

```python
#Generate the constants for the y-intercept 
xA = sm.add_constant(x1A)
xB = sm.add_constant(x1B)
xC = sm.add_constant(x1C)
xD = sm.add_constant(x1D)
xE = sm.add_constant(x1E)
xF = sm.add_constant(x1F)
xG = sm.add_constant(x1G) 
```

```python
#Model A Regression 
#Since the variables in the data frame are object types instead of numeric types you can set the parameters of the regression to float
results = sm.OLS(y.astype(float),xA.astype(float)).fit()
results.summary()
```

```python
#Model B Regression
results = sm.OLS(y.astype(float),xB.astype(float)).fit()
results.summary()
```

```python
#Model C Regression
results = sm.OLS(y.astype(float),xC.astype(float)).fit()
results.summary()
```

```python
#Model D Regression
results = sm.OLS(y.astype(float),xD.astype(float)).fit()
results.summary()
```

```python
#Model E Regression
results = sm.OLS(y.astype(float),xE.astype(float)).fit()
results.summary()
```

```python
#Model F Regression
results = sm.OLS(y.astype(float),xF.astype(float)).fit()
results.summary()
```

```python
#Model G Regression
results = sm.OLS(y.astype(float),xG.astype(float)).fit()
results.summary()
```
