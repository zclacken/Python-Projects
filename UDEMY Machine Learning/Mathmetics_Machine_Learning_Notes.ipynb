Question 1
Bank robber drives away at 150km/hr (2.5km/min) with a 5-minute head start
Sheriff drives after him at 180 km/hr (3km/min)
At what distance will the sheriff catch up with the robber? 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Use linspace function to to designate a start point, end point, and number of points between on the plot
t = np.linspace(0,40,1000) # start, finish, n points
```

```python
#Distance traveled by robber
d_r = 2.5 *t
```

```python
#Distance traveled by sheriff
d_s = 3*(t-5)
```

```python
#Plot equations to display the time and distance when their paths cross
fig,ax = plt.subplots()
plt.title("Sheriff Catches Up to Robber")
plt.xlabel('time(minutes)')
plt.ylabel('distance(Kilometers)')
ax.set_xlim([0,40])
ax.set_ylim([0,100])
ax.plot(t,d_r,c= 'red')
ax.plot(t,d_s,c = 'blue')
plt.axvline(x=30, color = 'purple',linestyle = '--')
_ = plt.axhline(y = 75, color = 'purple', linestyle = '--')
```
